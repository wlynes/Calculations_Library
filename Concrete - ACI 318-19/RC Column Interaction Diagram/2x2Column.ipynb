```python
# Import libraries
import math
import numpy as np
import matplotlib.pyplot as plt
from sectionproperties.pre.library.concrete_sections import concrete_column_section
```

```python
# User Input
fc = 3.5                             # ksi
fy = 53.7                            # ksi
b = 30                              # in
h = 30                             # in
dbar = 1.5                           # in
nbar_row = 2                         # number of bars top/bottom
cc = 2.5                             # distance from edge to center of rebar, in
```

```python
# Functions

def steel_props():
    AsTop = nbar_row * math.pi/4 * dbar**2
    AsBottom = AsTop
    E = 29000
    ey = fy / E
    ysTop = cc
    ysBottom = h - cc
    return AsTop, AsBottom, E, ey, ysTop, ysBottom

def conc_props():
    Ec = 57000 * math.sqrt(fc * 1000)
    if fc <= 4.:
        beta1 = 0.85
    elif fc < 8:
        beta1 = 0.85 - 0.05*(fc-4)
    else:
        beta1 = 0.65
    ec = 0.003
    Ag = b * h
    return Ec, beta1, ec, Ag

def safety_factor(e):
    if abs(e) <= ey:
        phi, classification = 0.65, 'CC'
    elif abs(e) < ey + 0.003:
        phi, classification = 0.65+0.25*(abs(e)-ey)/(0.003), 'Tr'
    else:
        phi, classification = 0.90, 'TC'
    return 1.5/phi, classification
    
def stress_strain():
    if c > ysBottom + dbar / 2:
        Acc = b * a - (AsTop - AsBottom)
        esBottom = ec / c * (c - ysBottom)
        fsBottom = min(fy, esBottom * E)
        esTop =ec / c * (c - ysTop)
        fsTop = min(fy, esTop * E)             
    elif c > ysTop + dbar / 2:
        Acc = b * a - AsTop
        esBottom = -1 * ec / c * (ysBottom - c)
        fsBottom = max(-fy, esBottom * E)  
        esTop =ec / c * (c - ysTop)
        fsTop = min(fy, esTop * E) 
    else:
        Acc = b * a
        esBottom = -1 * ec / c * (ysBottom - c)
        fsBottom = max(-fy, esBottom * E) 
        esTop = -1 * ec / c * (ysTop - c)
        fsTop = max(-fy, esTop * E) 
    return Acc, fsBottom, fsTop, esBottom
    
def calc_forces():
    Cc = 0.85 * fc * Acc
    FsTop = fsTop * AsTop
    FsBottom = fsBottom * AsBottom
    P = min(0.80*P0, Cc + FsTop + FsBottom)
    M = Cc*(h/2 - a/2) + FsTop*(h/2 - ysTop) + FsBottom*(h/2 - ysBottom)
    return P, M
```

```python
AsTop, AsBottom, E, ey, ysTop, ysBottom = steel_props()
Ec, beta1, ec, Ag = conc_props()

```

```python
# Initialize arrays
Pn = []
Mn = []
Pa = []
Ma = []
```

```python
# Iterate through depths of c, from bottom of section to top of section beginning with pure compression
P0 = 0.85 * fc * (Ag - (AsTop + AsBottom)) + fy * (AsTop + AsBottom)
Pn.append(0.80*P0)
Mn.append(0.)
Pa.append(0.80*P0/(1.50/0.65))
Ma.append(0.)

c = h

while c >= 0.:
    a = beta1 * c
    Acc, fsBottom, fsTop, esBottom = stress_strain()
    W, classification = safety_factor(esBottom)
    P, M = calc_forces()
    Pn.append(P)
    Mn.append(M/12)
    Pa.append(P/W)
    Ma.append((M/12)/W)
    c -= h/1000
```

```python
# Plot interaction diagram

fig, ax = plt.subplots()
ax.plot(Ma,Pa)
plt.title('Interaction Diagram')
plt.xlabel('Mn / \u03A9 (kip-ft)')
plt.ylabel('Pn / \u03A9 (kip)')
plt.grid(True)
plt.show()

# Plot section
geometry = concrete_column_section(b, h, cc-dbar/2, nbar_row, nbar_row, dbar, math.pi/4 * dbar**2, n_circle=20)
geometry.plot_geometry()
```
